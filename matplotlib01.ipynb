```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14,8))

fig.suptitle('Plot examples', fontsize=20)

# Scatter plot
mean = [20, 10]
cov = [[1,0.7],[0.7,1]]
x1, y1 = np.random.multivariate_normal(mean, cov, 1000).T
ax1.scatter(x1, y1, edgecolors='black')
ax1.set_title('Scatter plot example')

# Histogram
x2 = np.random.geometric(p=0.1, size=3000)
ax2.hist(x2, bins=40, color='red')
ax2.set_title('Hist plot example')

# Line plot
x3 = np.linspace(0.001, 6, 1000)
y_log = np.log(x3)
y_log2 = np.log(2*x3)
ax3.plot(x3, y_log, label='$log(x)$')
ax3.plot(x3, y_log2, label='$log(2x)$')
ax3.legend(loc='best')
ax3.set_title('Line plot example')

# Bar plot
size = 14
start_dt = np.datetime64('2021-01-01')
period = np.timedelta64(size, 'D')
x4 = np.arange(start_dt, start_dt+period).astype(str)
y4 = np.random.normal(loc=5, size=size)
ax4.tick_params(axis='x', labelrotation = 90)
ax4.bar(x4,y4)
ax4.set_title('Bar plot example')
```
